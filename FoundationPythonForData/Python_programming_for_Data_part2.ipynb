### 📌 Conditional Selection and Filtering in Pandas

In pandas, conditional selection is used to filter rows or columns based on specific conditions.

#### 🔹 Common Operators

* `==` → Equal to
* `>` → Greater than
* `<` → Less than
* `>=`, `<=` → Greater/Less than or equal to

#### 🔹 Combining Conditions

* `&` → Logical **AND** (both conditions must be true)
* `|` → Logical **OR** (at least one condition must be true)

> ⚠️ Always wrap each condition in parentheses when using `&` or `|`.

````python
df1.Name == "Adwoa"
````

````python
df1.loc[df1.Gender == 'Fm']
````

````python
df1.loc[(df1.Name == 'Adwoa') & (df1.Gender== 'Fm')]
````

````python
df1.loc[(df1.Name == 'Adwoa') | (df1.Gender== 'Fm')]
````

### GroupBy in Pandas

In data preparation for machine learning, each row represents a unique entity (e.g., a student, customer, or observation) with specific features (columns).

Sometimes, we want to analyze or transform data based on shared characteristics.  
For example, grouping all students with the same name or all samples from the same class.

In groupby, the column you group by becomes the index of the result, not the original rows.

Pandas provides the `groupby()` function to support this.

### 🔹 What is `groupby()`?

`groupby()` is used to group rows in a dataset based on the values of one or more columns.

It follows a three-step process:
1. **Split** the data into groups based on a column (e.g., "Name")
2. **Apply** a function to each group
3. **Combine** the results into a new structure

---

### 🔹 Example

```python
df.groupby("Name")["Score"].mean()

````python
import pandas as pd

data = {
    "Name": ["Ama", "Kofi", "Ama", "Yaw", "Kofi"],
    "City": ["Accra", "Kumasi", "Accra", "Tamale", "Kumasi"],
    "Score": [80, 75, 90, 60, 85]
}

df_group = pd.DataFrame(data)
print(df_group)
````

````python
df_group.groupby('Name')
````

from the code above  `df_group.groupby('Name')`

The dataframe is grouped as:

Ama →

   ` Name   City   Score`

    Ama    Accra  80

    Ama    Accra  90


Kofi →

    Name   City   Score

    Kofi   Kumasi 75

    Kofi   Kumasi 85


Yaw →
    Name   City   Score

    Yaw    Tamale 60

````python
df_group.groupby('Name').Score.min()
````

````python
df_group.groupby('Name').Score.max()
````

````python
df_group.groupby('Name').Score.mean()
````




`apply():` lets you run a custom function on each group, where each group is treated as a mini DataFrame. It is flexible and can return a single value, a Series, or even a DataFrame.

`transform():` method allows you to execute a function for each value of the DataFrame.


`agg():` Is used to compute summary statistics for each group, often reducing each group to a single value per function such as min, max

````python
df_group
````

````python
# used to impact the elements within a column in the dataset

df_group.groupby('Name').apply(lambda a: a.Score.iloc[-1])
````

````python
df_group.groupby('Name').Score.agg(['min','max','mean'])
````
